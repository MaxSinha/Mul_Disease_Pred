**IMPORTING THE DEPENDENCIES**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python
# loading diabetes dataset to pandas dataset
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# loading first 5 rows
diabetes_dataset.head()
```

```python
#Loading last 5 rows dataset
diabetes_dataset.tail()
```

```python
# Diabetes Dataset Shape

diabetes_dataset.shape
```

```python
# Checking info
diabetes_dataset.info()
```

```python
# Checking wheather any null values are present in this dataset or not
diabetes_dataset.isnull().sum()
```

```python
# getting the statistical measures of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

**0** ***--->*** **Represent Non Diabetic**  
**1** ***--->*** **Represent Diabetic**

```python
# getting average value of all columns
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Seperting Data and Labels

X=diabetes_dataset.drop(columns='Outcome',axis=1)
Y=diabetes_dataset['Outcome']
```

```python
# Printing X
X.head()
```

```python
# Printing Y
Y.head()
```

Data Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardlize_data= scaler.transform(X)
```

```python
print(standardlize_data)
```

```python
X=standardlize_data
Y=diabetes_dataset['Outcome']
```

# **SPLITTING DATA**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2 ,stratify=Y,random_state=2)
```

```python
# Checking train test split data
print(X.shape , X_train.shape , X_test.shape)
```

## **Training Model**

```python
classifier=svm.SVC(kernel='linear')
```

```python
# Training Support Vector Machine Classifier
classifier.fit(X_train,Y_train)
```

MODEL EVALUATION

Accuracy Score

```python
X_train_prediction = classifier.predict(X_train)
X_train_Accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Accuracy score of the training data : ', X_train_Accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
X_test_Accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('Accuracy score of the test data : ', X_test_Accuracy)
```

#**Making a Predictive System**

```python
input_data=(1,85,66,29,0,26.6,0.351,31) # Fill medical data

# Changing input data into numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)
print()

prediction = classifier.predict(std_data)
print(prediction)
print()

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
